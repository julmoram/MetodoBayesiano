#**Metodo Bayesiano en Marketing**

Usando el método bayesiano con distribución Beta, se calculó la probabilidad posterior de conversión de cada campaña. Mediante simulación Monte Carlo, se encontró que la campaña B tiene aproximadamente un 75% de probabilidad de ser superior a la campaña A. Por lo tanto, la empresa debería seleccionar la campaña B.

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos
clientes_A = 500
compras_A = 100

clientes_B = 500
compras_B = 150

# Fracasos
no_compras_A = clientes_A - compras_A
no_compras_B = clientes_B - compras_B

# Prior Beta(1,1)
alpha_prior = 1
beta_prior = 1

# Posterior
alpha_A = alpha_prior + compras_A
beta_A = beta_prior + no_compras_A

alpha_B = alpha_prior + compras_B
beta_B = beta_prior + no_compras_B

# Simulación Monte Carlo
simulaciones = 100000

theta_A = np.random.beta(alpha_A, beta_A, simulaciones)
theta_B = np.random.beta(alpha_B, beta_B, simulaciones)

# Probabilidad de que B sea mejor que A
prob_B_mejor = np.mean(theta_B > theta_A)
prob_A_mejor = np.mean(theta_A > theta_B)

print("Probabilidad de que Campaña A sea mejor:", prob_A_mejor)
print("Probabilidad de que Campaña B sea mejor:", prob_B_mejor)

# Gráfica
plt.hist(theta_A, bins=100, alpha=0.5, label="Campaña A")
plt.hist(theta_B, bins=100, alpha=0.5, label="Campaña B")
plt.legend()
plt.title("Distribuciones posteriores")
plt.show()
```
